```python
import pandas as pd
df = pd.read_csv('taiwan_covid_19_daily_reports_vaccine_city.csv')
df.head()
```

```python
vaccine_AZ = df[['City', 'Date' ,'Population', 'Daily vaccinations (AZ)', 'Total vaccinations (AZ)', 'Percentage (AZ,%)']]
vaccine_AZ.head()
```

```python
vaccine_AZ.count()
```

```python
vaccine_AZ_above18_city = vaccine_AZ.loc[vaccine_AZ['Percentage (AZ,%)'] > 18.0, ['City', 'Date', 'Percentage (AZ,%)']]
vaccine_AZ_above18_city
```

```python
AZ_high_city = vaccine_AZ_above18_city[['City', 'Percentage (AZ,%)']].groupby('City').mean().sort_values(by=['Percentage (AZ,%)'], ascending=False)
#將AZ施打率平均大於18%的資料group起來，並排序出來
AZ_high_city
```

```python
import matplotlib.pyplot as plt
AZ_high_city.plot()
```

```python
labels = vaccine_AZ_above18_city['City'].unique()
colors = ['lightblue','coral', 'red', 'yellow', 'green', 'blue']
plt.pie(AZ_high_city['Percentage (AZ,%)'], labels=labels, colors=colors, autopct='$'startangle=140)
plt.show()
```
