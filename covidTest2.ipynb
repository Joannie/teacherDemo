```python
vaccine_AZ = df[['City', 'Date' ,'Population', 'Daily vaccinations (AZ)', 'Total vaccinations (AZ)', 'Percentage (AZ,%)']]
vaccine_AZ.head()
```

```python
vaccine_AZ.count()

vaccine_AZ["City"].count()
```

```python
vaccine_AZ_above18_city = vaccine_AZ.loc[vaccine_AZ['Percentage (AZ,%)'] > 18.0, ['City', 'Date', 'Percentage (AZ,%)']]
vaccine_AZ_above18_city
```

```python
AZ_high_city = vaccine_AZ_above18_city[['City', 'Percentage (AZ,%)']].groupby('City').mean().sort_values(by=['Percentage (AZ,%)'], ascending=False)
#將AZ施打率平均大於18%的資料group起來，並排序出來
AZ_high_city
```

```python
import matplotlib.pyplot as plt

AZ_high_city.plot.bar()
plt.xlabel("City")
plt.ylabel("Percentage (AZ,%)")
plt.show()
```

```python
city_count = vaccine_AZ_above18_city["City"].value_counts()
city_count
```

```python
import matplotlib.pyplot as plt

plt.bar(city_count.index.values, city_count.values)
plt.xlabel("Drug Regimen")
plt.xticks(rotation=90)
plt.ylabel("Number of Mice")
plt.show()
```

```python
cities = vaccine_AZ_above18_city['City'].value_counts()
colors = ['lightblue','coral', 'red', 'yellow', 'green', 'pink']

cities.plot.pie(autopct="%1.1f%%", legend=False, colors = colors)
plt.title("AZ > 18%")
plt.show()
```
